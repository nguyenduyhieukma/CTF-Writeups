### guess_number

There exists small $(\epsilon_0, \epsilon_1, ..., \epsilon_{21}, )$ such that:
$$
t_0^{-1}(u_0 + \epsilon_0) = t_1^{-1}(u_1 + \epsilon_1) = t_2^{-1}(u_2 + \epsilon_2) = ... = t_{21}^{-1}(u_{21} + \epsilon_{21}) = \alpha \text{ (mod $p$)}
$$

This is equivalent to:
$$
\left\{
\begin{array}{ll}
t_1t_0^{-1}\epsilon_0 - \epsilon_1 = u_1 - t_1t_0^{-1}u_0 \\ 
t_2t_0^{-1}\epsilon_0 - \epsilon_2 = u_2 - t_2t_0^{-1}u_0 \\ 
... \\
t_{21}t_0^{-1}\epsilon_0 - \epsilon_{21} = u_{21} - t_{21}t_0^{-1}u_0 \\ 
\end{array} 
\right. \text{(all mod p)}
$$

Let $k_i = t_it_0^{-1}, c_i = u_i - k_i$, we have:
$$
\left\{
\begin{array}{ll}
1.\epsilon_0 = \epsilon_0 \\
k_1\epsilon_0 + l_1p = c_1 + \epsilon_1  \\ 
k_2\epsilon_0 + l_2p = c_2 + \epsilon_2  \\ 
... \\
k_{21}\epsilon_0 + l_{21}p = c_{21} + \epsilon_{21}\\ 
\end{array} 
\right. (l_i \in \mathbb{Z})
$$

Which means:
$$
\begin{bmatrix}
1 \\
k_1 \\
k_2 \\
... \\ 
k_{21}
\end{bmatrix}.\epsilon_0 +
\begin{bmatrix}
0 \\
p \\
0 \\
... \\ 
0
\end{bmatrix}.l_1 +
\begin{bmatrix}
0 \\
0 \\
p \\
... \\ 
0
\end{bmatrix}.l_2 +
... +
\begin{bmatrix}
0 \\
0 \\
0 \\
... \\ 
p
\end{bmatrix}.l_{21} =
\begin{bmatrix}
0 \\
c_1 \\
c_2 \\
... \\ 
c_{21}
\end{bmatrix} +
\begin{bmatrix}
\epsilon_0 \\
\epsilon_1 \\
\epsilon_2 \\
... \\ 
\epsilon_{21}
\end{bmatrix}
$$

This is a common lattice problem named [Closest vector problem (CVP)](https://en.wikipedia.org/wiki/Lattice_problem#Closest_vector_problem_.28CVP.29). In our situation, we need to find a linear combination with integer coefficients of those known column-vectors on the left-hand side, which results in a vector as close to $(0, c_1, c_2, ..., c_{21})$ as possible. The difference $(\epsilon_0, \epsilon_1, ..., \epsilon_{21})$ is what we really want to get. Since knowing $\epsilon_i$, calculating $t_i^{-1}(u_i + \epsilon_i)$ would give us $\alpha$.

To solve the problem above, we may use the LLL algorithm to reduce those column-vectors on the left to a "better" basis $B$ (which consists of short, nearly orthogonal vectors), temporarily ignore $(\epsilon_0, \epsilon_1, ..., \epsilon_{21})$, solve for $x$ in the equation $Bx = c$ with $c = (0, c_1, c_2, ..., c_{21})$, and finally round each component of $x$ to its nearest integer to obtain $\bar{x}$. The closest vector and the difference would be $B\bar{x}$ and $B\bar{x}-c$ respectively.
